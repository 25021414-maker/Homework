#### Bài 2: 
Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 

```python
a = float(input())
b = float(input())
PI = 3.14
khu_vui_choi = PI*((a//2)**2)
trong_cay = a*b-(khu_vui_choi)
print(f'{trong_cay:.2f}')
```

nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó

#### Bài 8: 
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh   
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng


```python
ten = input("Ten chu ho: ")
chi_so_truoc = float(input("Chi so thang truoc: "))
chi_so_nay = float(input("Chi so thang nay: "))
luong_dien = chi_so_nay - chi_so_truoc
tien_dien = 0

if luong_dien <= 50:
    tien_dien = luong_dien * 1984
elif luong_dien <= 100:
    tien_dien = 50 * 1984 + (luong_dien - 50) * 2050
elif luong_dien <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (luong_dien - 100) * 2380
elif luong_dien <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (luong_dien - 200) * 2998
elif luong_dien <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (luong_dien - 300) * 3350
else:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (luong_dien - 400) * 3460


tien_vat = tien_dien * 0.08
tong_tien = tien_dien + tien_vat


print(f"\n{'='*50}")
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {int(tong_tien)} ")
print(f"{'='*50}")
```
